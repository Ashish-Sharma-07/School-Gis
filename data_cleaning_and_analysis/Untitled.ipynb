```python
import csv
from functools import reduce
lst = dict()
with open('School_14-15/maharashtra/facilities.csv') as f1:
    next(f1)
    for line in csv.reader(f1):
        #print line[0]
        add = reduce(lambda x,y: int(x)+int(y) if x!=''and y!='' else int(x) if x!=''else int(y) if y!='' else 0 ,line[3:] )
        lst[line[0]] = add
```
